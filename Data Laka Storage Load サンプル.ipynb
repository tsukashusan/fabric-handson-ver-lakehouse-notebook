### 型の設定

## 型変換

```python
from pyspark.sql.types import IntegerType, TimestampType, DecimalType

shiresaki = shiresaki.withColumn("仕入先コード",shiresaki.仕入先コード.cast(IntegerType()))

zyuchu = zyuchu.withColumn("受注コード", zyuchu.受注コード.cast(IntegerType())) \
               .withColumn("得意先コード", zyuchu.得意先コード.cast(IntegerType())) \
               .withColumn("社員コード", zyuchu.社員コード.cast(IntegerType())) \
               .withColumn("受注日", zyuchu.受注日.cast(TimestampType())) \
               .withColumn("締切日", zyuchu.締切日.cast(TimestampType())) \
               .withColumn("出荷日", zyuchu.出荷日.cast(TimestampType())) \
               .withColumn("運送料", zyuchu.運送料.cast(DecimalType()))

zyuchumeisai = zyuchumeisai.withColumn("受注コード", zyuchumeisai.受注コード.cast(IntegerType())) \
                           .withColumn("商品コード", zyuchumeisai.商品コード.cast(IntegerType())) \
                           .withColumn("単価", zyuchumeisai.単価.cast(DecimalType())) \
                           .withColumn("数量", zyuchumeisai.数量.cast(IntegerType())) \
                           .withColumn("割引", zyuchumeisai.割引.cast(DecimalType()))

shohin = shohin.withColumn("商品コード", shohin.商品コード.cast(IntegerType())) \
               .withColumn("仕入先コード", shohin.仕入先コード.cast(IntegerType())) \
               .withColumn("区分コード", shohin.区分コード.cast(IntegerType())) \
               .withColumn("単価", shohin.単価.cast(DecimalType())) \
               .withColumn("在庫", shohin.在庫.cast(IntegerType()))


shohinkubun = shohinkubun.withColumn("区分コード", shohinkubun.区分コード.cast(IntegerType()))

tokuisaki = tokuisaki.withColumn("得意先コード", tokuisaki.得意先コード.cast(IntegerType()))

shain = shain.withColumn("社員コード", shain.社員コード.cast(IntegerType())) \
             .withColumn("誕生日", shain.誕生日.cast(TimestampType())) \
             .withColumn("入社日", shain.入社日.cast(TimestampType()))

unso = unso.withColumn("運送コード", unso.運送コード.cast(IntegerType()))
```

## 出力操作

### parquet ファイルへ書き出し

```python
shiresaki.write.mode('overwrite').parquet("Files/parquet/仕入先/")
zyuchu.write.mode('overwrite').parquet("Files/parquet/受注/")
zyuchumeisai.write.mode('overwrite').parquet("Files/parquet/受注明細/")
shohin.write.mode('overwrite').parquet("Files/parquet/商品/")
shohinkubun.write.mode('overwrite').parquet("Files/parquet/商品区分/")
tokuisaki.write.mode('overwrite').parquet("Files/parquet/得意先/")
shain.write.mode('overwrite').parquet("Files/parquet/社員/")
unso.write.mode('overwrite').parquet("Files/parquet/運送会社/")
todofuken.write.mode('overwrite').parquet("Files/parquet/都道府県/")
```

## 確認操作

### parquetファイルの読み出し

```python
parquet_shiresaki = spark.read.format("parquet").load("Files/parquet/仕入先/")
parquet_zyuchu = spark.read.format("parquet").load("Files/parquet/受注/")
parquet_zyuchumeisai = spark.read.format("parquet").load("Files/parquet/受注明細/")
parquet_shohin = spark.read.format("parquet").load("Files/parquet/商品/")
parquet_shohinkubun = spark.read.format("parquet").load("Files/parquet/商品区分/")
parquet_tokuisaki = spark.read.format("parquet").load("Files/parquet/得意先/")
parquet_shain = spark.read.format("parquet").load("Files/parquet/社員/")
parquet_unso = spark.read.format("parquet").load("Files/parquet/運送会社/")
parquet_todofuken = spark.read.format("parquet").load("Files/parquet/都道府県/")
```

```python
display(parquet_shiresaki)
```

### dela lake への書き出し

```python
parquet_shiresaki.write.mode("overwrite").format("delta").saveAsTable("vendors")
parquet_zyuchu.write.mode("overwrite").format("delta").saveAsTable("orders")
parquet_zyuchumeisai.write.mode("overwrite").format("delta").saveAsTable("order_details")
parquet_shohin.write.mode("overwrite").format("delta").saveAsTable("items")
parquet_shohinkubun.write.mode("overwrite").format("delta").saveAsTable("item_categories")
parquet_tokuisaki.write.mode("overwrite").format("delta").saveAsTable("customers")
parquet_shain.write.mode("overwrite").format("delta").saveAsTable("employees")
parquet_unso.write.mode("overwrite").format("delta").saveAsTable("shipping_companies")
parquet_todofuken.write.mode("overwrite").format("delta").saveAsTable("prefectures")
```
